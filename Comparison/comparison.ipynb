```python
import csv
import pytchat

processed_comment_datetimes = set()

# Create or open the CSV file for writing
csv_file_path = 'comparison.csv'
with open(csv_file_path, 'a', encoding='utf-8', newline='') as csv_file:
    # Create a CSV writer
    csv_writer = csv.writer(csv_file)

    # Fetch live chat messages and perform hate speech detection
    chat = pytchat.create(video_id="gCNeDWCI0vo")
    while chat.is_alive():
        for c in chat.get().items:
            comment_datetime = c.datetime
            author_name = c.author.name
            message = c.message

            # Check if comment has already been processed
            if comment_datetime in processed_comment_datetimes:
                continue

            print(f"{comment_datetime} [{author_name}]- {message}\n")

            # Write comment details into different columns
            csv_writer.writerow([comment_datetime, author_name, message])

            processed_comment_datetimes.add(comment_datetime)
```

### Hate Speech Detection Using Bloom Filter 

#### Setting up Bloom Filter

```python
bloom_filter = ScalableBloomFilter(mode=ScalableBloomFilter.SMALL_SET_GROWTH)

# Add hate speech terms to the filter from the file hurtlex_EN.tsv
with open("hurtlex_EN.tsv", "r") as file:
    for line in file:
        term = line.split("\t")[3]
        print(term)
        bloom_filter.add(term)
        
def detect_hate_speech(text):
    for term in text.split():
        if term in bloom_filter:
            return True
    return False
```

#### Measuring Time for Bloom Filter

```python
import time

csv_file_path = 'comparison.csv'
# open the CSV file for reading
with open(csv_file_path, 'r', encoding='utf-8', newline='') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)

    # Measure time taken for hate speech detection
    start_time = time.time()
    for row in csv_reader:
        comment_datetime = row[0]
        author_name = row[1]
        message = row[2]

        if detect_hate_speech(message):
            print(f"{comment_datetime} [{author_name}]- {message} - HATE SPEECH DETECTED\n")
    end_time = time.time()
    print(f"Time taken for hate speech detection: {end_time - start_time} seconds\n")




```

#### Measuring time for Pre-Trained Model

```python
classifier = pipeline("text-classification", model='IMSyPP/hate_speech_en')

# Set to store processed comment datetimes
processed_comment_datetimes = set()

# Open CSV file for writing
csv_file_path = 'comparison.csv'

# open the CSV file for reading
with open(csv_file_path, 'r', encoding='utf-8', newline='') as csv_file:
    # Create a CSV reader
    csv_reader = csv.reader(csv_file)

    # Measure time taken for hate speech detection
    start_time = time.time()
    for row in csv_reader:
        comment_datetime = row[0]
        author_name = row[1]
        message = row[2]

        # Check if comment has already been processed
        if comment_datetime in processed_comment_datetimes:
            continue

        # Perform hate speech detection
        result = classifier(message)
        if result[0]['label'] == 'LABEL_0':
            print(f"{comment_datetime} [{author_name}]- {message} - HATE SPEECH DETECTED\n")

        processed_comment_datetimes.add(comment_datetime)
    end_time = time.time()
    print(f"Time taken for hate speech detection: {end_time - start_time} seconds\n")




```
