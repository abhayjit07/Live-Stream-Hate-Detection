```python
from transformers import pipeline
import csv
import time

# Load the text classification pipeline
classifier = pipeline("text-classification", model='IMSyPP/hate_speech_en')

# Define the input and output CSV file paths
input_csv_path = 'Chat.csv'
output_csv_path = 'Updated_Chat.csv'

# Infinite loop to keep the script running
while True:
    # Open the input and output CSV files
    with open(input_csv_path, 'r', newline='', encoding='utf-8') as csvfile, \
            open(output_csv_path, 'a', newline='', encoding='utf-8') as output_csvfile:

        # Create CSV reader and writer
        reader = csv.reader(csvfile)
        writer = csv.writer(output_csvfile)

        # Write headers to the output CSV file
        headers = next(reader)
        writer.writerow(headers)

        # Iterate through rows in the input CSV file
        for row in reader:
            # Check if the label is 'hate'
            if row[3] == 'hate':
                # Perform hate speech detection
                result = classifier(row[2])

                # Update the label based on hate speech detection result
                if result[0]['label'] != 'LABEL_0' and result[0]['score'] > 0.5:
                    row[3] = 'hate'
                else:
                    row[3] = 'safe'

            # Write the updated row to the output CSV file
            writer.writerow(row)

    print(f"Hate speech detection results written to {output_csv_path}")

    # Sleep for a specified interval before checking the file again
    time.sleep(5)  # Adjust the sleep interval as needed
```
